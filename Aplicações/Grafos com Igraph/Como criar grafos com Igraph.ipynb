# Como criar grafos usando a biblioteca IGraph

Primeiramente precisamos importar a biblioteca em nosso projeto.

```python
from igraph import *
```

```python
# primeiro você cria o grafo
g = Graph()
# Depois você adiciona os vértices
g.add_vertices(4)
# por fim, adicione os edges
g.add_edges([(1,2),(2,3),(0,3)])
# plote
plot(g)
```

```python
g = Graph()  
g.add_vertices(4)  
g.add_edges([(1,2),(2,3),(0,3)])  
# nesse exemplo você pode adicionar nome aos vértices
plot(g, vertex_label=["A", "B", "C", "D"], vertex_color="white") 
```

```python
eG=Graph(directed=False)
eG.add_vertices(6)
eG.add_edges([(0,1),(1,2),(2,3),(3,4),(3,5),(5,3)])
# os pesos podem ser adicionados também
eG.es['weight']=[12,1,2,3,4,1]

layout = eG.layout("kk")

visual_style = {}
visual_style["vertex_size"] = 20
visual_style["vertex_label"] = ["a","b","c","d","e","f"]
visual_style["edge_width"] = eG.es['weight']
visual_style["bbox"] = (300, 300)


plot(eG, **visual_style)
```
